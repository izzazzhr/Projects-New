```python
mob20201 = pd.read_csv("/content/drive/MyDrive/Portfolio-Projects/Speed Test/mobile_year_2020_quarter_01.csv")
mob20202 = pd.read_csv("/content/drive/MyDrive/Portfolio-Projects/Speed Test/mobile_year_2020_quarter_02.csv")
mob20203 = pd.read_csv("/content/drive/MyDrive/Portfolio-Projects/Speed Test/mobile_year_2020_quarter_03.csv")
mob20204 = pd.read_csv("/content/drive/MyDrive/Portfolio-Projects/Speed Test/mobile_year_2020_quarter_04.csv")
mob20211 = pd.read_csv("/content/drive/MyDrive/Portfolio-Projects/Speed Test/mobile_year_2021_quarter_01.csv")
mob20212 = pd.read_csv("/content/drive/MyDrive/Portfolio-Projects/Speed Test/mobile_year_2021_quarter_02.csv")
mob20213 = pd.read_csv("/content/drive/MyDrive/Portfolio-Projects/Speed Test/mobile_year_2021_quarter_03.csv")
mob20214 = pd.read_csv("/content/drive/MyDrive/Portfolio-Projects/Speed Test/mobile_year_2021_quarter_04.csv")
mob20221 = pd.read_csv("/content/drive/MyDrive/Portfolio-Projects/Speed Test/mobile_year_2022_quarter_01.csv")
mob20222 = pd.read_csv("/content/drive/MyDrive/Portfolio-Projects/Speed Test/mobile_year_2022_quarter_02.csv")
mob20223 = pd.read_csv("/content/drive/MyDrive/Portfolio-Projects/Speed Test/mobile_year_2022_quarter_03.csv")
```

## Combine Datasets

```python
fix = pd.concat([fix20201, fix20202, fix20203, fix20204, fix20211, fix20212, fix20213, fix20214, fix20221, fix20222, fix20223], axis=0).reset_index()
mobile= pd.concat([mob20201, mob20202, mob20203, mob20204, mob20211, mob20212, mob20213, mob20214, mob20221, mob20222, mob20223], axis=0).reset_index()
```

```python
fix = fix.drop("index", axis=1)
mobile = mobile.drop("index", axis=1)
```

## Set Flag for Dashboard Filtering

```python
fix['Flag']='fixed'
mobile['Flag']='mobile'
```

## Get Last Quarter Value for Each Country for Comparison

```python
fix.head()
```

```python
# Step 1: Sort the DataFrame by Country, Year, and Quarter
fix.sort_values(['Name', 'Year', 'Quarter'], inplace=True)
```

```python
fix['LQ_Avg_Up'] = fix.groupby('Name')['Avg. Avg U Kbps'].shift()
fix['LQ_Avg_DL'] = fix.groupby('Name')['Avg. Avg D Kbps'].shift()
fix['LQ_Latency'] = fix.groupby('Name')['Avg Lat Ms'].shift()
fix['LQ_Devices'] = fix.groupby('Name')['Devices'].shift()
fix['LQ_Tests'] = fix.groupby('Name')['Tests'].shift()
fix['LQ_Rank_Up'] = fix.groupby('Name')['Rank Upload'].shift()
fix['LQ_Rank_DL'] = fix.groupby('Name')['Rank Download'].shift()
fix['LQ_Rank_Lat'] = fix.groupby('Name')['Rank Latency'].shift()
```

```python
fix = fix.sort_index()
fix.head()
```

```python
# Step 1: Sort the DataFrame by Country, Year, and Quarter
mobile.sort_values(['Name', 'Year', 'Quarter'], inplace=True)
```

```python
mobile['LQ_Avg_Up'] = mobile.groupby('Name')['Avg. Avg U Kbps'].shift()
mobile['LQ_Avg_DL'] = mobile.groupby('Name')['Avg. Avg D Kbps'].shift()
mobile['LQ_Latency'] = mobile.groupby('Name')['Avg Lat Ms'].shift()
mobile['LQ_Devices'] = mobile.groupby('Name')['Devices'].shift()
mobile['LQ_Tests'] = mobile.groupby('Name')['Tests'].shift()
mobile['LQ_Rank_Up'] = mobile.groupby('Name')['Rank Upload'].shift()
mobile['LQ_Rank_DL'] = mobile.groupby('Name')['Rank Download'].shift()
mobile['LQ_Rank_Lat'] = mobile.groupby('Name')['Rank Latency'].shift()
```

## Combine Mobile and Fixed Data

```python
data = pd.concat([fix, mobile], axis=0).reset_index()
```

```python
data = data.drop("index", axis=1)
data
```

```python
data['YearQuarter'] = data['Year'].astype(str) + 'Q' + data['Quarter'].astype(str)
```
